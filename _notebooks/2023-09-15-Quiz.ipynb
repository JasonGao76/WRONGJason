<h1>Binary to Decimal Quiz</h1>

<div id="quiz-container">
  <p id="question"></p>
  <input type="number" id="answer" min="0" max="31">
  <button onclick="checkAnswer()">Check Answer</button>
  <p id="result"></p>
  <button onclick="nextQuestion()">Next Question</button>
</div>

<script>
  // Quiz data
  const quizData = [
    "00000","00001","00010","00011","00100", "00101","00111","01000","01001","01010","01011","01100"
    // Add more binary numbers as needed
  ];

  let currentQuestionIndex = 0;

  // Function to initialize the quiz
  function initQuiz() {
    displayQuestion();
  }

  // Function to display the current question
  function displayQuestion() {
    const questionElement = document.getElementById("question");
    const currentQuestion = quizData[currentQuestionIndex];
    questionElement.textContent = `Convert the binary number ${currentQuestion} to decimal.`;
  }

  // Function to check the answer
  function checkAnswer() {
    const answerInput = document.getElementById("answer");
    const resultElement = document.getElementById("result");
    const userAnswer = parseInt(answerInput.value, 10);
    const correctAnswer = parseInt(quizData[currentQuestionIndex], 2);

    if (userAnswer === correctAnswer) {
      resultElement.textContent = "Correct!";
    } else {
      resultElement.textContent = "Incorrect. Please try again.";
    }
  }

  // Function to move to the next question
  function nextQuestion() {
    const answerInput = document.getElementById("answer");
    const resultElement = document.getElementById("result");

    // Reset input and result elements
    answerInput.value = "";
    resultElement.textContent = "";

    // Move to the next question
    currentQuestionIndex = (currentQuestionIndex + 1) % quizData.length;

    // Display the new question
    displayQuestion();
  }

  // Initialize the quiz
  initQuiz();
</script>
